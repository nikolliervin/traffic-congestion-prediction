```python

import cv2
import os
from datetime import datetime, timedelta


input_video_path = r"C:\Users\lilvinzivert\Desktop\TrafficVideos\001.mp4"


output_folder = r"C:\Users\lilvinzivert\Desktop\TrafficVideos\01"
os.makedirs(output_folder, exist_ok=True)


start_datetime = datetime(2023, 5, 5, 9, 0, 0)

time_interval = timedelta(seconds=30)

video_duration = 8 * 60

cap = cv2.VideoCapture(input_video_path)

current_time = 0
screenshot_count = 0

while current_time <= video_duration:
    ret, frame = cap.read()
    if not ret:
        break

    current_timestamp = start_datetime + timedelta(seconds=current_time)
    
    time_formatted = current_timestamp.strftime("%I-%M-%S %p")
    
    screenshot_filename = f"{current_timestamp.strftime('%Y-%m-%d')} {time_formatted}.jpg"
    screenshot_path = os.path.join(output_folder, screenshot_filename)
    cv2.imwrite(screenshot_path, frame)
    print(f"Saved screenshot: {screenshot_filename}")
    
    screenshot_count += 1

    if screenshot_count == 3:
        start_datetime += timedelta(days=1)
        screenshot_count = 0
    
    current_time += time_interval.total_seconds()
    cap.set(cv2.CAP_PROP_POS_MSEC, current_time * 1000) 

cap.release()
cv2.destroyAllWindows()
```

```python
import os
from ultralytics import YOLO
import pandas as pd

model = YOLO('yolov8m.pt')

image_dir = r'C:\Users\lilvinzivert\Desktop\TrafficVideos\01'

results_list = []

for filename in os.listdir(image_dir):
    if filename.endswith(('.jpg', '.png', '.jpeg')):  
        image_path = os.path.join(image_dir, filename)
        
        results = model.predict(image_path)
        result = results[0] 
        vehicle_count = 0
        for box in result.boxes:
            class_id = int(box.cls[0].item())  
            class_name = result.names[class_id] 
            if class_name in ['car', 'truck', 'bus', 'ambulance']:
                vehicle_count += 1
        
        results_list.append({'id': len(results_list) + 1, 'name of image': filename, 'vehicle_detected': vehicle_count})

df = pd.DataFrame(results_list)

output_excel_path = 'detection_results.xlsx'
df.to_excel(output_excel_path, index=False, engine='openpyxl')

print(f"Detection results saved to {output_excel_path}")
```
